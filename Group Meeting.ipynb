## <center> Polynomial Chaos Expansion

### Cameron-Martin theorem guarantees the homogeneous chaos expansion converges to any processes with finie second-order moments.

$$
y(t,\zeta) = \sum_{i=0}^{\infty} y_i(t)\Phi_i(\zeta)   \ \  \Rightarrow \ \  y(t,\zeta) = \sum_{i=0}^P y_i(t)\Phi_i(\zeta)
$$
where $\{\Phi_i(\cdot)\}$ is a group of orthogonal polynomials

Orthogonality of polynomials here means $\int_S \Phi_i(x)\Phi_j(x)w(x)\ dx = \delta_{ij}$

### PC applied to ODE
$$
\frac{dy(t)}{dt} = -ky, \ \ \ y(0)=\hat{y}_0
$$

### PC applied to PDE example

$$
\frac{\partial u}{\partial t} = k \frac{\partial^2 u}{\partial x^2}, \ \ \ x\in[0,\pi], \ t>0
$$

$$ u(0,t) = u(\pi,t)=0 $$
$$ u(x,0) = \sin (x) , \ \ x\in[0,\pi]$$

u(x,t,k) = e^{-kt}\sin x

## <center> Karhunen-Loeve Expansion
 

(**Karhunen-Loeve**) Let $X: D\times\Omega \to \mathbb{R}$ be a centered, mean-square continuous stochastic process with $X\in L^2(D \times \Omega)$. There exists a basis $\{e_i\}$ of $L^2(D)$ such that for all $t\in D$,
$$
X_t(w) = X(t,w) = \sum_{i=1}^{\infty} x_i(w) e_i(t) \ \ \  x_i(w) = \int_D X_t(w) e_i(t)\ dt
$$
and 
1. $E[x_i] = 0$


2. $E[x_ix_j] = \delta_{ij}\lambda_j$


3. $\text{Var}[x_i] = \lambda_i$


***Compare with PCE:*** Deterministic term vs. randomness term

### Why are people interested in it?  

- In the deterministic case, discretization of the domain has a physical appeal. But the domain in the stochastic case does not have a physical meaning that permits a sensible discretization.

- Continuous stochastic process $\{X_t\}_{t\in D}$ where $t$ belongs to some continuous uncountable set $D$. If there exists a dense subset $A$ of $D$, we will be able to approximate the process by $\{X_t\}_{t\in A}$. Still a large number of random variables is required

- One very important result is the spectral representation of random process, which is stated as

\begin{align*}
w(x,\theta) &= \int g(x)\ d\mu(\theta)\\
C_{ww}(x_1, x_2) &= \int g(x_1)g(x_2)\ \langle d\mu_1(\theta) \ d\mu_2(\theta)\rangle
\end{align*}
where $d\mu(\theta)$ is termed orthogonal stochastic measure. This representation involves differentials of random functions  in an infinite dimensional space.

- KL expansion (Stochastic version of Fourier expansion)
$$
X(t,w) = \sum_{i=1}^{\infty} x_i(w) e_i(t)
$$
A stochastic process is represented by a denumerable set of random variables.

$\{e_i\}$ is a basis of a Hilbert space, then we can treat $x_i$ as the projection of $X(t,w)$ on the $e_i$ direction.

If we trancate the above equation, we only need a finite number of random variables to approximate the stochastic process.

### <font color = red> A stochastic process is mean-square continuous if
$$
\lim_{\epsilon\to 0} E[(X_{t+\epsilon} - X_t)^2] = 0
$$
 
$\{X_t\}_{t\in [a,b]}$ is mean-square continuous if and only if $R_X$ is continuous on $[a,b]\times[a,b]$

### <font color = red> Hilbert-Schmidt Integral Operator
    
$X: D\times\Omega \to \mathbb{R}$ be a centered, mean-square continuous stochastic process with $X\in L^2(D\times \Omega)$, define the integral operator $K: L^2(D)\to L^2(D)$ by
$$
[Ku](s) = \int_D k(s,t)u(t)\ dt, \ \  k(s,t) = R_X(s,t)
$$
Then, $K$ is compact, positive and self-adjoint.

### <font color = red> Spectral theorem for compact self-adjoint operators
    
$H$: (real or complex) Hilbert space, $T: H\to H$ is a compact self-adjoint operator. Then $H$ has an orthonormal basis $\{e_i\}$ of eigenvectors of $T$ corresponding to eigenvalues $\lambda_i$.

Since $L^{2}(D)$ is Hilbert space, from above, we see that there exists orthomormal basis $\{e_i\}$ of $K$.

### <font color = red> Mercer's Theorem
    
Let $k: D\times D\to \mathbb{R}$ be a continuous function, where $D=[a,b]\subset\mathbb{R}$. Suppose further that the corresponding Hilbert-Schmidt operator $K: L^2(D)\to L^2(D)$ is positive. If $\{\lambda_i\}$ and $\{e_i\}$ are the eigenvalues and eigenvectors of $K$, then for all $s, t\in D$,
$$
k(s,t) = \sum_i \lambda_i e_i(s)e_i(t)
$$
where convergence is absolute and uniform on $D\times D$.

By Mercer's Theorem, we are able to get the spectral decomposition of the autocorrelation function
$$
R_X^N(s,t) = \sum_{i=1}^N \lambda_i e_i(s) e_i(t)
$$

### A classical example
Gaussian random field $X$, variance $\sigma^2$, autocorrelation function is
$$
R_X(s,t) = \sigma^2 exp(-\frac{|s-t|}{L_c})
$$

![approx cov](approx_cov.png)

The KL expansion can be implemented for the random coefficients in the operator equation (Random operator); it cannot be implemented for the solution process, since its covariance function and therefore the corresponding eigenfunctions(Fredholm Eq) are not known

###  Error Minimizing Properties (Need Mercer's Thm)
$$
\lim_{N\to\infty} E[(X_t - \sum_{i=1}^N x_i e_i(t))^2] = 0 
$$

(**Normalized Karhunen-Loeve**) Let $X: D\times\Omega \to \mathbb{R}$ be a centered, mean-square continuous stochastic process with $X\in L^2(D\times \Omega)$. There exists a basis $\{e_i\}$ of $L^2(D)$ such that for all $t\in D$,
$$
X(t,w) = \sum_{i=1}^{\infty} \sqrt{\lambda_i} \xi_i(w) e_i(t)  \ \ \text{ in } L^{2}(\Omega)
$$
where $\xi_i$ are centered mutually uncorrelated random variables with unit variance and are given by
$$
\xi_i(w) = \frac{1}{\sqrt{\lambda_i}} \int_D X_t(w) e_i(t)\ dt
$$

Note: Gaussian process, $\xi_i$ are independent $N(0,1)$ and the series representation converges almost surely.


### Some known results

#### The Brownian motion
Covariance function of the Brownian motion is $K(s,t) = \min(s,t)$, the Karhunen-Loeve expansion for the Brownian motion is
$$
W(t) = \sqrt{2}\sum_{i\geq 1} \frac{2}{(2i-1)\pi} \sin[(i-\frac{1}{2})\pi t]\xi_i
$$
where $\{\xi_i\}$ are i.i.d. Gaussian process with mean zero and variance 1.

```python
a = 0.5
c = 1
```

```python
# 1-wtan(0.5w)=0
w_odd = [1.307, 6.585, 12.723, 18.955]
# x+tan(0.5w)=0
w_even = [3.673, 9.632, 15.834, 22.082]
```

```python
def eigenfunc1(n,x):
    w = w_odd[n]
    return np.cos(w*x)/np.sqrt(a+np.sin(2*w*a)/(2*w))

def eigenfunc2(n,x):
    w = w_even[n]
    return np.sin(w*x)/np.sqrt(a-np.sin(2*w*a)/(2*w))
```

```python
x = np.linspace(-a, a, 100)
y1 = eigenfunc1(0,x)    #??
y3 = eigenfunc1(1,x)    #third
y2 = eigenfunc2(0,x)     #second
y4 = eigenfunc2(1,x)     #Fourth
```

```python
plt.plot(x,y1, label='1')
plt.plot(x,y2,'--', label='2')
plt.plot(x,y3, '-.', label='3')
plt.plot(x,y4, label='4')
plt.ylabel('EIGENFUNCTION '+ '$f_n(x)$')
plt.legend()
```

```python
def eigenvalue(c,x):
    return 2*c/(x**2+c**2)
```

```python
b = [0.1, 0.5, 1.0, 2.0, 5.0, 10.0]
```

```python
eigen_matrix = np.zeros((6, 8)) #each row lists the eigenvalues for a certain b
```

```python
for i in range(len(b)):
    c = 1/b[i]
    for j in range(8):
        if j%2 ==0:
            x = w_odd[j//2]
        else:
            x = w_even[(j-1)//2]
        eigen_matrix[i,j] = eigenvalue(c,x)    
```

```python
x = [1,2,3,4,5,6,7,8]
for i in range(6):
    y = eigen_matrix[i,:]
    plt.plot(x,y, label=b[i])
    plt.legend()
plt.ylabel('EIGENVALUE '+ '$\lambda_n$');   
```

```python
x = np.linspace(-0.5,0.5,100)
y = np.linspace(-0.5,0.5,100)
xx, yy = np.meshgrid(x,y)
z = np.exp(-np.abs(xx-yy))

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(xx,yy,z,cmap=cm.jet)
plt.colorbar(surf,fraction=0.026, pad=0.0)
```

```python
c = 1
def approx(n,x1,x2):
    y = 0
    for i in range(n):
        if i%2 == 0:
            y += eigen_matrix[2,i]*eigenfunc1(i//2, x1)*eigenfunc1(i//2, x2)
        else:
            y += eigen_matrix[2,i]*eigenfunc1((i-1)//2, x1)*eigenfunc1((i-1)//2, x2)
    return y
```

```python
x = np.linspace(-0.5,0.5,100)
y = np.linspace(-0.5,0.5,100)
xx, yy = np.meshgrid(x,y)
z = approx(4,xx,yy)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(xx,yy,z,cmap=cm.jet)
plt.colorbar(surf,fraction=0.026, pad=0.0)
```
